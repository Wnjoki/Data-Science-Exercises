```python
pip install PyPDF2
import pandas as pd
import spacy
from PyPDF2 import PdfReader
```

Preprocessing

```python

def Function_to_extract_text_from_pdf(file_path):
    with open(file_path, "rb") as file:
        reader = PdfReader(file)
        text = ""
        for page in reader.pages:
            text += page.extract_text()
    return text

pdf_file = "/content/Speech.pdf"
text = Function_to_extract_text_from_pdf(pdf_file) 

# Remove leading and trailing spaces
import re
text = re.sub(r'^\s+|\s+$', '', text)

# Remove multiple consecutive spaces
speech = re.sub(r'\s+', ' ', text)

#convert to lower case
speech = speech.lower()
print(speech)


```

```python
# Tokenization
nlp = spacy.load("en_core_web_sm")

speech_doc = nlp(speech)
tokens = [token.text for token in speech_doc]

# Stop words removal
filtered_tokens = [token.text for token in speech_doc if not token.is_stop]

# Print filtered tokens after stop words removal
print("Tokens after Stop Words Removal:")
print(filtered_tokens)
print()

# Lemmatization
lemmas = [token.lemma_ for token in speech_doc]

# Print lemmas
print("Lemmas:")
print(lemmas)


print("Contents of Preprocessed text:")
print(filtered_tokens)

```

```python
#print_pos_tags
speech_doc = nlp(speech)
pos_tags = [(token.text, token.pos_) for token in speech_doc]
print("List of Part-of-speech tags:")
for token, pos in pos_tags:
    print(f"{token}: {pos}")



```

```python
#print_named_entities(text_processed)
named_entities = [(entity.text, entity.label_) for entity in speech_doc.ents]

# Print named entities
print("\nNamed Entities:")
for entity, label in named_entities:
    print(f"{entity}: {label}") 

  



```
