```python

import wikipediaapi
from sklearn.feature_extraction.text import TfidfVectorizer

```

```python
# Create a Wikipedia API object
wiki_wiki = wikipediaapi.Wikipedia('en')
```

```python
#social document titles
document_titles = [
    "Income Inequality",
    "Poverty",
    "Gender inequality",
    "Environmental justice"
]
```

```python
# get the text content of the documents
document_content = []
for title in document_titles:
    page = wiki_wiki.page(title)
    if page.exists():
        document_content.append(page.text)
        file_name = f"{title}.txt"
        with open(file_name, "w", encoding="utf-8") as file:
          file.write(page.text)
          print(f"Document '{title}' saved as '{file_name}'.")
    else:
            print(f"Page '{title}' does not exist on Wikipedia.")

            
```

```python
# Initialize the TF-IDF vectorizer
vectorizer = TfidfVectorizer(stop_words = 'english')

# Compute the TF-IDF scores
tfidf_scores = vectorizer.fit_transform(document_content)

# Get the feature names (terms)
feature_names = vectorizer.get_feature_names_out()
```

```python
#reducing the documents score to 500

top_count = 500
#Print the TF-IDF scores for each document
for i, doc_idx in enumerate(tfidf_scores):
    doc_scores = doc_idx.toarray()[0]
    print(f"\nTF-IDF scores for document: {document_titles[i]}")
    top_term_indices = doc_scores.argsort()[-top_count:][::-1]  # Get indices of top terms
    for term_idx in top_term_indices:
        term = feature_names[term_idx]
        score = doc_scores[term_idx]
        print(f"{term}: {score:.4f}")
```
